```python
im = np.zeros((32, 32)) 
im[:, 11:21] = 1

plt.imshow(im)
```

```python
corr = []
for i in range(0, 360, 10):
    m = transform.rotate(im, i)
    c = (m - m.mean())*(im - im.mean())
    c = c.mean()
    corr.append(c)
    
```
